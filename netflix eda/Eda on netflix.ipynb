```python
#  Find maximum duration / Seasons of Movies
df['duration_min_Only']=df['duration']
movies_duration=df[df['type']=='Movie'].groupby('duration_min_Only').size().reset_index().rename(columns={0:'No. of movie'}).sort_values(by='duration_min_Only',ascending=False)

movies_duration['duration_min_Only']=movies_duration['duration_min_Only'].str.replace("min",'').astype(int)
movies_duration.sort_values(by='duration_min_Only',ascending=False)
```

```python
# Find maximum duration / Seasons of TV show

df['duration_season_Only']=df['duration']
tvshow_season=df[df['type']=='TV Show'].groupby(['duration_season_Only']).size().reset_index().rename(columns={0:'No. of TV Show'}).sort_values(by='duration_season_Only')
tvshow_season['duration_season_Only']=tvshow_season['duration_season_Only'].str.replace("Seasons",'')
tvshow_season['duration_season_Only']=tvshow_season['duration_season_Only'].str.replace("Season",'').astype(int)
tvshow_season.sort_values(by='duration_season_Only',ascending=False)
```
